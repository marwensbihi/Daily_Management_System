Time Series Analysis
Time Series Plots

```python
date_column_name = 'Date'  # Change this to the correct column name

# Convert the date column to datetime with error handling
merged_final_data[date_column_name] = pd.to_datetime(merged_final_data[date_column_name], errors='coerce')

# Remove rows with NaT in the date column
merged_final_data = merged_final_data.dropna(subset=[date_column_name])

# Remove '%' and convert to float
merged_final_data['Tx. Abs '] = merged_final_data['Tx. Abs '].str.rstrip('%').astype(float)
merged_final_data['Tx. Polyvalence'] = merged_final_data['Tx. Polyvalence'].str.rstrip('%').astype(float)

# Plot the Time Series
plt.figure(figsize=(15, 7))
merged_final_data.set_index(date_column_name)[['Tx. Abs ', 'Tx. Polyvalence']].plot()
plt.title('Time Series Plot of Absenteeism Rate and Polyvalence Rate')
plt.xlabel('Date')
plt.ylabel('Values')
plt.show()
```

 Performance Comparison
Radar Charts


```python
categories = list(merged_final_data.columns[1:]) # Adjust based on your KPI columns
N = len(categories)
values = merged_final_data.loc[0].drop('Date').values.flatten().tolist()
values += values[:1]
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]

plt.figure(figsize=(10, 10))
ax = plt.subplot(111, polar=True)
plt.xticks(angles[:-1], categories, color='grey', size=8)
ax.plot(angles, values)
ax.fill(angles, values, 'b', alpha=0.1)
plt.title('Radar Chart of KPIs')
plt.show()
```

Bubble Charts

```python
import matplotlib.pyplot as plt

# Example DataFrame
data = {
    'QtéProduite_x': [100, 200, 150, 250],
    'Qtérebut.Int': [10, 20, 15, 25],
    'Tx. Abs': ['3.8%', '4.2%', '3.5%', '5.0%']
}

merged_final_data = pd.DataFrame(data)

# Remove '%' and convert to float
merged_final_data['Tx. Abs'] = merged_final_data['Tx. Abs'].str.rstrip('%').astype(float)

# Ensure all columns are of the same length
assert len(merged_final_data['QtéProduite_x']) == len(merged_final_data['Qtérebut.Int']) == len(merged_final_data['Tx. Abs']), "Columns must be of the same length"

# Plot the Bubble Chart
plt.figure(figsize=(15, 7))
plt.scatter(
    merged_final_data['QtéProduite_x'], 
    merged_final_data['Qtérebut.Int'], 
    s=merged_final_data['Tx. Abs'] * 100,  # Use the corrected 'Tx. Abs' column
    alpha=0.5
)
plt.title('Bubble Chart of Produced Quantity, Internal Scrap Quantity, and Absenteeism Rate')
plt.xlabel('Produced Quantity')
plt.ylabel('Internal Scrap Quantity')
plt.show()
```
